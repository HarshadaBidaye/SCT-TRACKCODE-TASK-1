```python
#Task 1 : To complete create a distribution of a bar chart and histogram
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Create an Bar plot based on age and gender distribution
data={
    'Gender':['Male','Female','Male','Female','Male','Female','Male','Female'],
    'Age':[10,15,7,12,8,10,12,15],
}
df = pd.DataFrame(data)
print(df)

plt.figure(figure=(10,6))
sns.barplot(x='Age', hue='Gender',data=df, palette='Set2')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.title('Distribution of Data')

plt.show()

```

```python
sns.set(style="whitegrid")

# Create a histogram of age distribution by gender
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Gender', multiple='stack',kde=True, bins=10, palette='husl')

# Add titles and labels
plt.title("Age and Gender Distribution", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Gender", fontsize=12)

# Showplot
plt.show()

```
